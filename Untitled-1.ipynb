1. TENTUKAN LIBRARY YANG DIGUNAKAN

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2. LOAD DATASET

```python
Banjir_dataset = pd.read_csv('Data Banjir.csv')
```

```python
Banjir_dataset.head()
```

```python
Banjir_dataset.shape
```

```python
Banjir_dataset['Terjadi Banjir'].value_counts()
```

```python
# memisahkan data dan label
X = Banjir_dataset.drop(columns='Terjadi Banjir', axis=1)
Y = Banjir_dataset['Terjadi Banjir']
```

```python
print(X)
```

```python
print(Y)
```

4. STANDARISASI DATA

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = Banjir_dataset['Terjadi Banjir']
```

```python
print(X)
print(Y)
```

5. MEMISAHKAN DATA TRAINING DAN DATA TESTING

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

6. MEMBUAT DATA LATIH MENGGUNAKAN ALGORITMA SVM

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

7. MEMBUAT MODEL EVALUASI UNTUK MENGUKUR TINGKAT AKURASI

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Akurasi data training : ', training_data_accuracy)
```

```python
X_text_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_text_prediction, Y_test)
```

```python
print('Akurasi data testing : ', test_data_accuracy)
```

8. MEMBUAT MODEL PREDIKSI

```python
input_data = (3,2020,291,19.9,31.6,25.8,46,98,73.93)

input_data_as_numpy_array = np.array(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('Tidak akan terjadi banjir')
else:
    print('Akan terjadi banjir')
```
